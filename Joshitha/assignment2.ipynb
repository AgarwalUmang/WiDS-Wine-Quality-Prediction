```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data=pd.read_csv("winequality-red.csv")
```

```python
print(data.dtypes)
```

```python
for col in data.columns:
  count=(data[col]==0).sum()
  print("No of 0's in",col,"=",count)
```

```python
plt.plot(data["quality"],data["alcohol"])
plt.title("Line plot of wine contents")
plt.ylabel("Quality of wine")
plt.xlabel("Alcohol % of wine")
plt.show()
```

```python
data["antilog_pH"]=np.power(10,data["pH"])
print(data)
```
